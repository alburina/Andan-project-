# Предварительная обработка

```python
import pandas as pd
```

```python
df = pd.read_csv('movies_data_1.0.csv')
print(df.shape)
df
```

```python
df.info()
```

Для начала удалим пустые и почти пусты столбцы и нерелевантные для нашего исследования, такие как "premiere.bluray", "premiere.dvd", 'premiere.digital', "distributors.distributor", "distributors.distributorRelease", 'fees.russia.value', 'fees.russia.currency',
       'fees.usa.value', 'fees.usa.currency' а также "type", т.к. мы брали данные только по фильмамм'fees.russia', 'fees.usa'

```python
print("Столбцы в DataFrame:", df.columns) 
```

```python
columns_to_drop = [
    "rating.await", "premiere.country", 'votes.await', "premiere.cinema", 'premiere.digital', 'fees.russia.value', 'fees.russia.currency',
    'fees.usa.value', 'fees.usa.currency', "type",
    "fees.russia", "fees.usa", "fees.world", "premiere.bluray", "premiere.dvd", 'distributors.distributor', 'distributors.distributorRelease'
]

df = df.drop(columns=columns_to_drop, errors="ignore")
print(df.shape)
df
```

Проверим наличие пропусков


```python
df.isna().sum()
```

Удалим пропуски в категорях "длина фильма" и "жанр"

```python
df = df.dropna(subset=["movieLength"])
df = df.dropna(subset=["genres"])
df.shape
```

Теперь взгяленим на столбцы с инфорцмацией о бюджете фильма и кассовых сборов. 

```python
print(df['fees.world.currency'].value_counts())
```

Можем заметить что все данные о кассовых сборах в мирепредставленны в $, следовательно этот столбец можно удалить

```python
columns_to_drop = [
    "fees.world.currency"
]
df = df.drop(columns=columns_to_drop, errors="ignore")
```

```python
df['budget.currency'].value_counts()
```

Бюджет фильмов представлен в 19 разных валютах, но многие из них всетречаются несколько раз, поэтому мы их удалим. А оставшиеся данные представленные в 5-и валютах пререведем в доллары по нынешнему курсу.

```python
	
df.loc[df['budget.currency'] == '€', 'budget.value'] *= 1.09
df.loc[df['budget.currency'] == '₽', 'budget.value'] *= 0.012
df.loc[df['budget.currency'] == '£', 'budget.value'] *= 1.30
df.loc[df['budget.currency'] == 'FRF', 'budget.value'] *= 1.14

df = df[
    (df['budget.currency'] == '€') |
    (df['budget.currency'] == '$') |
    (df['budget.currency'] == '₽') |
    (df['budget.currency'] == '£') |
    (df['budget.currency'].isnull())
]

columns_to_drop = [
    "budget.currency"
]

df = df.drop(columns=columns_to_drop, errors="ignore")
```

Нулевые значения кассового сбора и бюджета заполним средним значением по столбцу

```python
df['budget.value'] = df['budget.value'].fillna(0).round().astype(int)
df['fees.world.value'] = df['fees.world.value'].fillna(0).round().astype(int)
df['budget.value'].value_counts()
```

```python
df.loc[df['budget.value'] == 0, 'budget.value'] = round(df.loc[df['budget.value'] != 0, 'budget.value'].mean())
df.loc[df['fees.world.value'] == 0, 'fees.world.value'] = round(df.loc[df['fees.world.value'] != 0, 'fees.world.value'].mean())
```

Посмотри на пропуски в датах премьер в России и в мире

```python
print((df['premiere.russia'].isnull() & df['premiere.world'].notnull()).sum())
print((df['premiere.russia'].notnull() & df['premiere.world'].isnull()).sum())
print((df['premiere.russia'].isnull() & df['premiere.world'].isnull()).sum())
```

Пропуски в столбцах с датой премьеры в России можно заполнить данными из столбца с мировой датой премьеры и, аналогично, пропуски дат мирового проката заполнить данными из столбца с датой премьеры в России. Оставшиеся 52 стороки, где нет данных обоих премьер стоит удалить.

```python
df['premiere.russia'] = df['premiere.russia'].fillna(df['premiere.world'])
df['premiere.world'] = df['premiere.world'].fillna(df['premiere.russia'])
df = df.dropna(subset=["premiere.world"])
df = df.dropna(subset=["premiere.russia"])
```

```python
import re
df['genres'] = df['genres'].apply(lambda x: re.findall(r"'name': '([^']*)'", str(x)))
```
