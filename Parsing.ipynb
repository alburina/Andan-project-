# Сбор Данных

Для начала спарсим данные о 5000 фильмах с Кинопоиска.

```python
import requests
import json
import time

headers = {"X-API-KEY": "3VYHQQR-XJ8MCCK-MJVY232-H2CZ775"}

def get_movies():
    all_movies = []
    page = 1
    total_needed = 5000
    movies_per_request = 250  
    while len(all_movies) < total_needed:
        params = {
            'selectFields': 'name movieLength genres rating type votes year budget fees distributors premiere',
            'type': 'movie',
            'sortField': 'votes.kp',
            'sortType': '-1',
            'limit': movies_per_request,
            'page': page  
        }

        response = requests.get('https://api.kinopoisk.dev/v1.3/movie', headers=headers, params=params)
        data = response.json()
        movies = data.get('docs', [])
        all_movies.extend(movies)
        print(f"Загружено: {len(all_movies)} фильмов")

        page += 1
        time.sleep(1)  

    return all_movies

movies = get_movies()

# Сохранение в файл
with open('data.json', 'w', encoding='utf-8') as file:
    json.dump(movies, file, indent=4, ensure_ascii=False)

print(f"Всего сохранено фильмов: {len(movies)}")
```
